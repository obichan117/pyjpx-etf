```python
import pyjpx_etf as etf
```

## ETF holdings as a DataFrame

```python
# Get holdings as a DataFrame
e = etf.ETF("1306")
df = e.to_dataframe()
df.head(10)
```

## ETF metadata

```python
print(f"Name: {e.info.name}")
print(f"Date: {e.info.date}")
print(f"Cash Component: {e.info.cash_component:,.0f}")
print(f"Shares Outstanding: {e.info.shares_outstanding:,}")
print(f"Number of holdings: {len(e.holdings)}")
```

## Top holdings by weight

```python
df.sort_values("weight", ascending=False).head(10)
```
